```python
import os
import pandas as pd
import json
import time
import re
import glob
import bs4 as bs4
from requests_html import HTMLSession
from tqdm import tqdm
```

```python
wd = "D:/Mestrado/2020-2/Curso DS/Recommender System - YT/get and clean data/"
saving_dir = wd+"raw_data/"
url = "https://www.youtube.com{link}"
```

```python
# parse source code using beautifulsoup to extract useful informations
with open(wd+"parsed_video_info.json", "w+") as output:
    for video_file in tqdm(sorted(glob.glob(saving_dir+"/video*")), desc="working on video files"):
        with open(video_file, "r+", encoding='utf8') as inp:
            page_html = inp.read()
            parsed = bs4.BeautifulSoup(page_html, 'html.parser')

            class_watch = parsed.find_all(attrs={"class": re.compile(r"watch")})
            id_watch = parsed.find_all(attrs={"id": re.compile(r"watch")})
            channel = parsed.find_all("a", attrs={"href": re.compile(r"channel")})
            meta = parsed.find_all("meta")
            like_ratio = parsed.find_all("div", attrs={"id":"tooltip"})

            data = dict()

            for e in class_watch:
                colname = "_".join(e['class'])
                if "clearfix" in colname or re.match(r"^css-https.*", colname):
                    continue
                data[colname] = e.text.strip()

            for e in id_watch:
                colname = e['id']
                # if "clearfix" in colname:
                #     continue
                # print(colname)
                data[colname] = e.text.strip()

            for e in meta:
                colname = e.get('name')
                if colname is None:
                    colname = e.get('property')
                if colname is None:
                    colname = e.get('itemprop')
                if colname is not None:
                    data[colname] = e['content']

            for e in like_ratio:
                lr_candidate = e.text.strip().split()
                if len(lr_candidate) == 3:
                    if lr_candidate[1] == '/':
                        try:
                            data['likes'] = int(lr_candidate[0].replace('.', ''))
                            data['dislikes'] = int(lr_candidate[2].replace('.', ''))
                        except:
                            print('not a like ratio')

            for link_num, e in enumerate(channel):
                data["channel_link_{}".format(link_num)] = e['href']
            output.write("{}\n".format(json.dumps(data)))
```
